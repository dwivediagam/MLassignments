**Question 1** : Use the whole D2z.txt as training set. Use Euclidean distance (i.e. A = I). Visualize the predictions of 1NN on a 2D grid [−2 : 0.1 : 2]^2. That is, you should produce test points whose first feature goes over −2, −1.9, −1.8, . . . , 1.9, 2, so does the second feature independent of the first feature. You should overlay the training set in the plot, just make sure we can tell which points are training, which are grid.


---



```python
import numpy as np

import matplotlib.pyplot as plt

train = np.loadtxt('D2z.txt')

x = np.arange(-2, 2.1, 0.1) # np.arange returns an array of evenly spaced values within a specified interval (start, stop, step_size)
y = np.arange(-2, 2.1, 0.1)
X, Y = np.meshgrid(x, y)

test = np.column_stack((X.ravel(), Y.ravel()))

from scipy.spatial.distance import cdist

distances = cdist(test, train[:, :-1]) # exclude the last column of train (which contains the class labels)

nn_indices = np.argmin(distances, axis=1)

nn_labels = train[nn_indices, -1]

plt.scatter(train[train[:, -1] == 1][:, 0], train[train[:, -1] == 1][:, 1], color='blue')
plt.scatter(train[train[:, -1] == 0][:, 0], train[train[:, -1] == 0][:, 1], color='red')
plt.scatter(test[:, 1], test[:, 0], c=nn_labels, cmap='bwr', alpha=0.2)
plt.show()
```

**Question 3** : Implement logistic regression (from scratch). Use gradient descent (refer to question 6 from part 1)
to find the optimal parameters. You may need to tune your learning rate to find a good optimum. Run 5-fold
cross validation. Report accuracy, precision, and recall in each fold.


---



```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class LogisticRegression:

  def sigmoid(self, val):
    return 1.0/(1.0 + np.exp(-val))

  def estimate(self, w, X):

    y_hat = self.sigmoid(np.dot(X, w))

    y_hat = np.where(y_hat > 0.5, 1, 0)

    return y_hat

  def model(self, X, y):

    rows = X.shape[0]
    cols = X.shape[1]

    w = np.zeros((cols, 1))

    lr = 0.0001

    for i in range(0, 1000):

      y_hat = self.estimate(w, X)

      w = w - lr * self.gradient(X, y, y_hat)

    return w, y_hat

  def gradient(self, X, y, y_hat):
    
    m = X.shape[0]
    
    # Gradient of loss w.r.t weights.
    dw = (1/m)*np.dot(X.T, (y_hat - y))

    return dw

  def accuracy(self, y_hat, y):
    accuracy = np.sum(y == y_hat) / len(y)
    return accuracy

  def __init__(self):
    
    data = pd.read_csv("emails.csv")

    X = data.iloc[:, 1:-1]

    y = data[['Prediction']].copy()

    m = X.shape[0]

    # Implement 5-fold cross validation
    folds = [(0,1000), (1000,2000), (2000,3000), (3000,4000), (4000,5000)]
    for fold in folds:
      # Split the data into training and test sets for this fold
      test_start, test_end = fold
      X_test = X[test_start:test_end]
      y_test = y[test_start:test_end]
      X_train = pd.concat([X[:test_start], X[test_end:]])
      y_train = pd.concat([y[:test_start], y[test_end:]])

      w, y_hat = self.model(X_train, y_train)

      acc = self.accuracy(y_hat, y_train)

      y_hat_test = self.estimate(w, X_test)

      acc_test = self.accuracy(y_hat_test, y_test)

      # Compute the accuracy, precision, and recall
      tp = np.sum(np.logical_and(y_test == 1, y_hat_test == 1))
      fp = np.sum(np.logical_and(y_test == 0, y_hat_test == 1))
      tn = np.sum(np.logical_and(y_test == 0, y_hat_test == 0))
      fn = np.sum(np.logical_and(y_test == 1, y_hat_test == 0))
      accuracy.append((tp + tn) / len(y_test))
      precision.append(tp / (tp + fp))
      recall.append(tp / (tp + fn))

accuracy = []
precision = []
recall = []

l = LogisticRegression()

for i in range(0, len(accuracy)):
  print("Fold number " + str(i+1))
  print(*accuracy[i])
  print(*precision[i])
  print(*recall[i])
  print("\n")
```

**Question 4** : Run 5-fold cross validation with kNN varying k (k=1, 3, 5, 7, 10). Plot the average accuracy versus
k, and list the average accuracy of each case.



---



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('emails.csv')

X = data.iloc[:, 1:-1]
y = data[['Prediction']].copy()

# Define range of k values
k_values = [1, 3, 5, 7, 10]

# List to store average accuracy for each k value
avg_acc = []

for k in k_values:

    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Perform 5-fold cross validation
    scores = cross_val_score(knn, X, y, cv=5)
    
    # Compute average accuracy
    avg_acc.append(np.mean(scores))

# Plot average accuracy vs k
plt.plot(k_values, avg_acc, marker='o')
plt.xlabel('k')
plt.ylabel('Average Accuracy')
plt.title('kNN Cross Validation')
plt.grid()
plt.show()

# Print average accuracy for each k value
for i in range(len(k_values)):
    print("k =", k_values[i], " Average Accuracy =", avg_acc[i])
```

**Question 5** : Use a single training/test setting. Train kNN (k=5) and logistic regression on the training set, and draw ROC curves based on the test set.



---



```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('emails.csv')

# Split the data into training and test sets
X = data.iloc[:, 1:-1]

y = data[['Prediction']].copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1000)

# Train kNN with k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict the probabilities for the test set
knn_probs = knn.predict_proba(X_test)[:, 1]

# Train logistic regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predict the probabilities for the test set
logreg_probs = logreg.predict_proba(X_test)[:, 1]

# Calculate the TPR and FPR for various thresholds
knn_fpr, knn_tpr, knn_thresholds = roc_curve(y_test, knn_probs)
knn_roc_auc = auc(knn_fpr, knn_tpr)

logreg_fpr, logreg_tpr, logreg_thresholds = roc_curve(y_test, logreg_probs)
logreg_roc_auc = auc(logreg_fpr, logreg_tpr)

# Plot the ROC curves
plt.figure()
plt.plot(knn_fpr, knn_tpr, color='blue', lw=2, label='kNN (AUC = %0.2f)' % knn_roc_auc)
plt.plot(logreg_fpr, logreg_tpr, color='darkorange', lw=2, label='Logistic Regression (AUC = %0.2f)' % logreg_roc_auc)
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
