## Plotting a ROC curve (Part 1 - Question 5a)

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Let '+' class be 1 and '-' class be 0

y_true = [1, 1, 0, 1, 1, 0, 1, 1, 0, 0]
y_pred = [0.95, 0.85, 0.8, 0.7, 0.55, 0.45, 0.4, 0.3, 0.2, 0.1]

fpr, tpr, thresholds = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='green', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='orange', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

## Programming  Part 1 - Visualize the D2z.txt

```python
import numpy as np


train = np.loadtxt('hw3Data/D2z.txt')
x_1 = np.arange(-2, 2.1, 0.1) 
x_2 = np.arange(-2, 2.1, 0.1)
X1, X2 = np.meshgrid(x_1, x_2)

test = np.column_stack((X1.ravel(), X2.ravel()))

from scipy.spatial.distance import cdist
dist = cdist(test, train[:, :-1], metric='euclidean') 

nn_indices = np.argmin(dist, axis=1)

nn_labels = train[nn_indices, -1]
import matplotlib.pyplot as plt

plt.scatter(train[train[:, -1] == 0][:, 0], train[train[:, -1] == 0][:, 1], color='green')
plt.scatter(train[train[:, -1] == 1][:, 0], train[train[:, -1] == 1][:, 1], color='red')
plt.scatter(test[:, 1], test[:, 0], c=nn_labels, cmap='RdYlGn', alpha=0.2)
plt.legend(["Train class 0", "Train class 1"], loc ="lower right") 
plt.show()
```

## Programming Spam filter Logistic Regression

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations

    def sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))

    def fit(self, X, y):
        self.theta = np.zeros((X.shape[1],1))  # Initialize weights
        m = X.shape[0]  # Number of training samples

        for i in range(self.num_iterations):
            z = np.dot(X, self.theta)
            h = self.sigmoid(z)
            h = h.reshape((len(h), 1))
#             print(X.T.shape, self.theta.shape, y.shape)
            gradient = np.dot(X.T, (h - y)) / m
            self.theta -= self.learning_rate * gradient

    def predict(self, X):
        z = np.dot(X, self.theta)
        y_hat = np.where(z > 0.5, 1, 0)
        return y_hat

if __name__ == "__main__":
    accuracy = []
    precision = []
    recall = []
    data = pd.read_csv("hw3Data/emails.csv")

    X = data.iloc[:, 1:-1]

    y = data[['Prediction']].copy()

    # Create and train the logistic regression model for different folds
    
    folds = [(0,1000), (1000,2000), (2000,3000), (3000,4000), (4000,5000)]
    for fold in folds:
        # Split the data into training and test sets for this fold
        test_start, test_end = fold
        X_test = X[test_start:test_end]
        y_test = y[test_start:test_end]
        X_train = pd.concat([X[:test_start], X[test_end:]])
        y_train = pd.concat([y[:test_start], y[test_end:]])
#         X_train = np.column_stack((np.ones(X_train.shape[0]), X_train))
        model = LogisticRegression()
        model.fit(X_train, y_train)
        
        # Predict
#         test_data = np.column_stack((np.ones(X_test.shape[0]), X_test))
        y_hat_test = model.predict(X_test)
        
#         print("Predictions:", predictions)
        
        # Compute the accuracy, precision, and recall
        tp = np.sum(np.logical_and(y_test == 1, y_hat_test == 1))
        fp = np.sum(np.logical_and(y_test == 0, y_hat_test == 1))
        tn = np.sum(np.logical_and(y_test == 0, y_hat_test == 0))
        fn = np.sum(np.logical_and(y_test == 1, y_hat_test == 0))
        acc = (tp + tn) / len(y_test)
        prec = tp / (tp + fp)
        rec = tp / (tp + fn)
        accuracy.append(acc)
        precision.append(prec)
        recall.append(rec)
    for i in range(0, len(accuracy)):
        print("Fold number " + str(i+1))
        print(accuracy[i])
        print(precision[i])
        print(recall[i])
```

## ROC comparison between kNN and Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
data = pd.read_csv('hw3data/emails.csv')

X = data.iloc[:, 1:-1]
y = data[['Prediction']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1000)

lreg = LogisticRegression()
lreg.fit(X_train, y_train)
lreg_pred_probs = lreg.predict_proba(X_test)[:, 1]
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred_probs = knn.predict_proba(X_test)[:, 1]

lreg_fpr, lreg_tpr, lreg_thresholds = roc_curve(y_test, lreg_pred_probs)
lreg_roc_auc = auc(lreg_fpr, lreg_tpr)
knn_fpr, knn_tpr, knn_thresholds = roc_curve(y_test, knn_pred_probs)
knn_roc_auc = auc(knn_fpr, knn_tpr)

plt.figure()
plt.plot(knn_fpr, knn_tpr, color='lightblue', lw=2, label='kNN (AUC = %0.3f)' % knn_roc_auc)
plt.plot(lreg_fpr, lreg_tpr, color='darkorange', lw=2, label='Logistic Regression (AUC = %0.3f)' % lreg_roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve comparison between kNN and Logistic Regression')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
